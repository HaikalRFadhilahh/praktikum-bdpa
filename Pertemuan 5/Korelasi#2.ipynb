## Import Library

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from pandas.plotting import scatter_matrix
```

## Import Dataset

```python
raw_data = pd.read_csv('./automobileEDA.csv')
raw_data
```

## Melihat DATASET

```python
raw_data.head(10)
```

```python
raw_data.tail()
```

## Mencari Nilai Korelasi

```python
# Mencari Korelasi menggunakan Metode Pearson
raw_data.corr()
```

```python
# Mencari Korelasi menggunakan metode kendall
raw_data.corr(method='kendall')
```

```python
# Mencari Korelasi Menggunakan metode Spearman
raw_data.corr(method='spearman')
```

## Matriks Korelasi

```python
plt.clf()

# mengatur ukuran gambar/plot
plt.rcParams['figure.dpi'] = 100

plt.figure()
plt.matshow(raw_data.corr())
plt.title('Plot correlation matriks dengan .matshow', size=14)
plt.tight_layout()
plt.show()

plt.figure()
sns.heatmap(raw_data.corr(), annot=True)
plt.title('Plot correlation matriks dengan sns.heatmap', size=14)
plt.tight_layout()
plt.show()
```

## Grouped Box Plot

```python
plt.clf()

plt.figure()
# boxplot biasa tanpa pengelompokkan
raw_data.boxplot(rot=90)
plt.title('Boxplot tanpa pengelompokkan', size=14)
plt.tight_layout()
plt.show()

plt.figure()
# box plot dengan pengelompokkan dilakukan oleh kolom 'Make'
raw_data.boxplot(by='make')
plt.tight_layout()
plt.show()
```

## Grupped Histogram

```python
plt.clf()

plt.figure()
raw_data[raw_data['aspiration'] == 'std'].hist()
plt.tight_layout()
plt.show()

plt.figure()
raw_data[raw_data['aspiration'] == 'turbo'].hist()
plt.tight_layout()
plt.show()
```

## Hex Bin Plot

```python
plt.clf()

plt.figure()
raw_data.plot.hexbin(x='compression-ratio', y='horsepower', gridsize=25, rot=90)
plt.tight_layout()
plt.show()
```

## Scatter Matrix Plot

```python
plt.clf()

_, ax = plt.subplots(1, 1, figsize=(10,10))
scatter_matrix(raw_data, ax=ax)
plt.show()
```
