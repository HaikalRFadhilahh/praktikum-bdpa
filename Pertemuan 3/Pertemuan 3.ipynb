```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
import warnings
```

```python
df = pd.read_csv('./fiba.csv');
df.head(1)
```

```python
df.shape
```

```python
df.columns
```

```python
df.describe()
```

```python
df_like = df.sort_values(by=['likeCount'],ascending=False);
```

```python
df_like_stats = df_like.iloc[:10];
df_comment = df.sort_values(by=['commentCount'],ascending=False).head(10)
```

## Bar Chart Comment Counting

```python
plt.figure(figsize=(12,7))
plt.bar(df_like_stats['title'],df_like_stats['commentCount']);
plt.title('Jumlah Like Di Video Terbanyak')
plt.xlabel('Judul Video')
plt.ylabel('Jumlah Like')
plt.xticks(rotation=90,ha='left')
warnings. filterwarnings("ignore")
plt.show()
```

## Line Chart Comment Counting

```python
plt.figure(figsize=(12,7))
plt.plot(df_comment['title'],df_comment['commentCount'],marker='o', color='r', label='line with marker');
plt.title('Jumlah Comment Di Video Terbanyak')
plt.xlabel('Judul Video')
plt.ylabel('Jumlah Comment')
plt.xticks(rotation=90,ha='left')
warnings. filterwarnings("ignore")
plt.show()
```
